```python
import pandas as pd
import selenium
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager

driver_path = print(ChromeDriverManager().install())  
driver = webdriver.Chrome(driver_path)
driver.get("https://www.youtube.com/results?search_query=mister+duncan")
```

```python
# Attendre un peu pour que la page charge
time.sleep(5)

# Scroller pour charger plus de vidÃ©os (facultatif)
driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
time.sleep(5)

# Extraire les informations des vidÃ©os
videos = driver.find_elements(By.XPATH, '//ytd-video-renderer')

# Listes pour stocker les donnÃ©es
video_types = []
titles = []
links = []
views = []
durations = []
upload_dates = []
```

```python
for video in videos:
    # Type de vidÃ©o (vÃ©rifier si c'est une vidÃ©o ou un live, etc.)
    video_type = "Video"
    
    # Extraire le titre et le lien
    try:
        title_element = video.find_element(By.XPATH, './/a[@id="video-title"]')
        title = title_element.text
        link = title_element.get_attribute('href')
    except:
        title = "N/A"
        link = "N/A"

    # Extraire le nombre de vues
    try:
        view_element = video.find_element(By.XPATH, './/span[contains(@class, "view-count")]')
        view_count = view_element.text if view_element else "N/A"
    except:
        view_count = "N/A"
    
    # Extraire la durÃ©e
    try:
        duration_element = video.find_element(By.XPATH, './/span[@class="style-scope ytd-thumbnail-overlay-time-status-renderer"]')
        duration = duration_element.text.strip()
    except:
        duration = "N/A"
    
    # Extraire la date de mise en ligne
    try:
        date_element = video.find_element(By.XPATH, './/div[@id="metadata-line"]/span[2]')
        upload_date = date_element.text if date_element else "N/A"
    except:
        upload_date = "N/A"

    # Ajouter les donnÃ©es Ã  la liste
    video_types.append(video_type)
    titles.append(title)
    links.append(link)
    views.append(view_count)
    durations.append(duration)
    upload_dates.append(upload_date)

# Fermer le navigateur
driver.quit()
```

```python
# CrÃ©er un DataFrame Pandas
df = pd.DataFrame({
    'Type de vidÃ©o': video_types,
    'Titre': titles,
    'Lien': links,
    'Nombre de vues': views,
    'DurÃ©e': durations,
    'Date de mise en ligne': upload_dates
})

import pandas as pd

# Affichage du DataFrame
print(df)

# Sauvegarder dans un fichier CSV
df.to_csv('resultats_videos_youtube.csv', index=False)
```
